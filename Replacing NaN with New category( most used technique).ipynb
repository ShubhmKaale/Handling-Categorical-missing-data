```python
list =['BsmtQual','FireplaceQu','GarageType','SalePrice']
df1 = df[list]
```

```python
df1.head()
```

```python
def impute_nan(dframe, variable):
    dframe[variable+"_newvar"]= np.where(dframe[variable].isnull(),"Missing", dframe[variable])
```

```python
for feature in ['BsmtQual','FireplaceQu','GarageType']:
    impute_nan(df1,feature)
                
```
